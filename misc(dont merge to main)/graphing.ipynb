```python
import matplotlib.pyplot as plt
import numpy as np
import sys
sys.path.insert(1, '/Users/angelapeng/Github/Group-1-Amoeba')
import matplotlib.patches as mpatches
import pickle
import loggerOutput
color_list =["#3eb991","#6ecadc","#e9a820","#e01563","#edb196","#1f94ac","#ae9a6a","#ccb8a6","#343a44"]
```

# Importing Data

```python
def read_pickle(file_name):
    data = None
    lables .append(file_name)
    with open("/Users/angelapeng/Github/Group-1-Amoeba/output_coord/"+file_name+".pickle", 'rb')as f:
    
        data = pickle.load(f)
    f.close()
    return data
```

```python
datas =[]
lables = []
if loggerOutput.comb_formation:
    datas .append( read_pickle("comb_formation"))
if loggerOutput.movable_location:
    datas .append( read_pickle("movable_location"))
if loggerOutput.periphery:
    datas .append( read_pickle("periphery"))
if loggerOutput.extend:
    datas .append( read_pickle("extend"))
if loggerOutput.retract:
    datas .append( read_pickle("retract"))
    
```

# Graphing the cells

```python
def graph(datas,):
    fig, ax = plt.subplots()
    i = 0
    colors = []
    for data in datas:
        x, y = zip(*data)
        colors.append(color_list[i])
        ax.scatter(x,y,color = color_list[i])
        i +=1


    ax.set_xticks(np.arange(30, 60, 5))
    ax.set_yticks(np.arange(0, 100, 5))
    lc = zip(lables,colors)
    handles = [mpatches.Patch(color=colors, label=lables) for lables, colors in lc]
    a = plt.legend(handles, lables, ncol=1, bbox_to_anchor=(1, 1))
```

```python
graph(datas)
```
